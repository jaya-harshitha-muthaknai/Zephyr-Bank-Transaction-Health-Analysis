```python
import pandas as pd
```

```python
import sqlite3
```

```python
customer = pd.read_csv("../data/dim_customer.csv")
merchant = pd.read_csv("../data/dim_merchant_category.csv")
transaction_type = pd.read_csv("../data/dim_transaction_type.csv")
transactions = pd.read_csv("../data/fact_transactions_updated.csv")
```

```python
# creating sql database
conn = sqlite3.connect(':memory:')
```

```python
customer.to_sql('customer', conn, index=False)
merchant.to_sql('merchant', conn, index=False)
transaction_type.to_sql('transaction_type', conn, index=False)
transactions.to_sql('transactions', conn, index=False)
```

```python
# 1 Are Starter customers using High-Risk merchant 
#categories more frequently than Premium customers?
query = """ SELECT
    c.customer_segment,
    COUNT(*) AS high_risk_transactions
FROM transactions t
JOIN customer c
    ON t.customer_id = c.customer_id
JOIN merchant m
    ON t.merchant_category_id = m.merchant_category_id
WHERE m.risk_flag = 'High'
GROUP BY c.customer_segment
ORDER BY high_risk_transactions DESC;"""
```

```python
high_risk_segment = pd.read_sql_query(query, conn) 
#Standard customers have the highest usage of High-Risk merchant categories, with 50 transactions recorded.
```

```python
high_risk_segment 
```

```python
# 2. Which customer segment generates the highest fee 
#revenue AFTER excluding fraud and failures?
query = """   SELECT
    c.customer_segment,
    SUM(tt.typical_fee_gbp) AS expected_revenue,
    SUM(t.fee_charged_gbp) AS actual_revenue,
    SUM(tt.typical_fee_gbp) - SUM(t.fee_charged_gbp) AS revenue_leakage
FROM transactions t
JOIN customer c
    ON t.customer_id = c.customer_id
JOIN transaction_type tt
    ON t.transaction_type_id = tt.transaction_type_id
GROUP BY c.customer_segment
ORDER BY revenue_leakage DESC;"""
```

```python

fee_revenue_by_segment = pd.read_sql_query(query, conn)
```

```python
fee_revenue_by_segment
#Business customers outperformed revenue expectations, 
#whereas Standard customers contributed the highest revenue shortfall.
```

```python
# 3. Are high-risk merchants actually profitable?
query = """SELECT
    m.risk_flag,
    SUM(t.amount_gbp) AS transaction_value,
    SUM(t.fee_charged_gbp) AS revenue,
    SUM(
        CASE
            WHEN t.is_flagged_fraud = TRUE
            THEN t.amount_gbp
            ELSE 0
        END
    ) AS fraud_exposure
FROM transactions t
JOIN merchant m
    ON t.merchant_category_id = m.merchant_category_id
GROUP BY m.risk_flag;"""
```

```python
high_risk_profit =  pd.read_sql_query(query, conn)
```

```python
high_risk_profit
#High-risk merchants generated low revenue but high fraud exposure, 
#indicating poor profitability.
```

```python
#4. Which region has the highest fraud rate relative to transaction volume?
query ="""  SELECT
    c.region,
    COUNT(*) AS transactions,
    SUM(
        CASE
            WHEN t.is_flagged_fraud = TRUE
            THEN 1
            ELSE 0
        END
    ) AS fraud_count,
    ROUND(
        100.0 *
        SUM(
            CASE
                WHEN t.is_flagged_fraud = TRUE
                THEN 1
                ELSE 0
            END
        ) / COUNT(*),
        2
    ) AS fraud_rate_pct
FROM transactions t
JOIN customer c
    ON t.customer_id = c.customer_id
GROUP BY c.region
ORDER BY fraud_rate_pct DESC; """
```

```python
region_high_fraud = pd.read_sql_query(query, conn)
```

```python
region_high_fraud 
#Fraud activity was heavily concentrated in London (60%) and the Midlands (33%),
#while no fraud transactions were observed in other regions during the analysis period.
```

```python
#5. Which transaction types create the largest revenue leakage?
query = """ SELECT
    tt.type_name,
    SUM(tt.typical_fee_gbp) AS expected_fee,
    SUM(t.fee_charged_gbp) AS actual_fee,
    SUM(tt.typical_fee_gbp) -
    SUM(t.fee_charged_gbp) AS leakage
FROM transactions t
JOIN transaction_type tt
    ON t.transaction_type_id =
       tt.transaction_type_id
GROUP BY tt.type_name
ORDER BY leakage DESC; """
```

```python
largest_revenue_leak = pd.read_sql_query(query, conn)
```

```python
largest_revenue_leak
#International Transfers showed the highest revenue leakage (£399.61), 
#meaning the bank collected significantly less fee revenue than expected.
```

```python
# 6.Which merchant categories have more fraud than their risk classification predicts?
query = """  SELECT
    m.category_name,
    m.risk_flag,
    COUNT(*) AS transactions,
    SUM(
        CASE
            WHEN t.is_flagged_fraud = TRUE
            THEN 1
            ELSE 0
        END
    ) AS fraud_count,
    ROUND(
        100.0 *
        SUM(
            CASE
                WHEN t.is_flagged_fraud = TRUE
                THEN 1
                ELSE 0
            END
        ) / COUNT(*),
        2
    ) AS fraud_rate
FROM transactions t
JOIN merchant m
ON t.merchant_category_id =
   m.merchant_category_id
GROUP BY
    m.category_name,
    m.risk_flag
ORDER BY fraud_rate DESC;"""
```

```python
merchant_fraud_risk = pd.read_sql_query(query, conn)
```

```python
merchant_fraud_risk 
# Fuel had the highest fraud rate (31.33%), making it the most fraud-prone merchant category.
```

```python
#7.How much fraud exposure exists within the bank's established customer base?
query = """  SELECT
    'Established Customers' AS customer_group,
    COUNT(*) AS transactions,
    SUM(
        CASE
            WHEN t.is_flagged_fraud = 1
            THEN 1
            ELSE 0
        END
    ) AS fraud_count,
    ROUND(
        100.0 *
        SUM(
            CASE
                WHEN t.is_flagged_fraud = 1
                THEN 1
                ELSE 0
            END
        ) / COUNT(*),
        2
    ) AS fraud_rate_pct
FROM transactions t
JOIN customer c
    ON t.customer_id = c.customer_id;"""
```

```python
Established_Customers = pd.read_sql_query(query, conn)
```

```python
Established_Customers
#Established customers (Customers whose accounts have been open for a long time.) generated 300 fraud transactions, 
#resulting in a 20% fraud rate across the customer base.
```

```python
#8. Which channel-device combination creates the highest fraud risk?
query = """ SELECT
    tt.channel,
    t.device_type,
    COUNT(*) AS transactions,
    SUM(
        CASE
            WHEN t.is_flagged_fraud = TRUE
            THEN 1
            ELSE 0
        END
    ) AS fraud_count
FROM transactions t
JOIN transaction_type tt
ON t.transaction_type_id =
   tt.transaction_type_id
GROUP BY
    tt.channel,
    t.device_type
ORDER BY fraud_count DESC;"""
```

```python
channel_fraud = pd.read_sql_query(query, conn)
```

```python
channel_fraud
#Automated transactions with missing device information (NaN) recorded the highest fraud risk, 
#driven entirely by Standing Orders, Salary Credits, and Loan Repayments.
#The Automated channel recorded the highest fraud activity,
#particularly in transactions where device information was unavailable.
```

```python
#9. Which customers generate both high revenue and high fraud exposure?
query = """  SELECT
    t.customer_id, 
    SUM(t.fee_charged_gbp) AS revenue,
    SUM(
        CASE
            WHEN t.is_flagged_fraud = TRUE
            THEN t.amount_gbp
            ELSE 0
        END
    ) AS fraud_exposure
FROM transactions t
GROUP BY t.customer_id
ORDER BY fraud_exposure DESC;"""
```

```python
customer_revenue_fraud = pd.read_sql_query(query, conn)
```

```python
customer_revenue_fraud
#Customer 20 emerged as a significant risk outlier, generating over £1.1M in fraud exposure
# through repeated high-value reversed transactions flagged as fraud.
```

```python
#10. Is international business profitable after accounting for fraud?
query = """ SELECT
    tt.is_domestic,
    SUM(t.fee_charged_gbp) AS revenue,
    SUM(
        CASE
            WHEN t.is_flagged_fraud = TRUE
            THEN t.amount_gbp
            ELSE 0
        END
    ) AS fraud_loss,
    SUM(t.fee_charged_gbp)
    -
    SUM(
        CASE
            WHEN t.is_flagged_fraud = TRUE
            THEN t.amount_gbp
            ELSE 0
        END
    ) AS net_profit_indicator
FROM transactions t
JOIN transaction_type tt
ON t.transaction_type_id =
   tt.transaction_type_id
GROUP BY tt.is_domestic;"""
```

```python
international_business_profit = pd.read_sql_query(query, conn)
```

```python
international_business_profit 
#Neither domestic nor international transactions were profitable, 
#as fraud losses significantly exceeded the fee revenue generated.
```
